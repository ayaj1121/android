```python
import pandas as pd,matplotlib.pyplot as plt,numpy as np
pos_salary=pd.read_csv("C:\\Users\\memon\Documents\\python\\decision and randomforest\\Position_Salaries.csv")
X=pos_salary.iloc[:,1:2].values
y=pos_salary.iloc[:,2].values
from sklearn.tree import DecisionTreeRegressor
regresser=DecisionTreeRegressor(random_state=0)
regresser.fit(X,y)
y_pred=regresser.predict([[6.5]])
X_grid=np.arange(min(X),max(X),0.01)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color="red")
plt.plot(X_grid,regresser.predict(X_grid),color="green")
plt.show()
print(regresser.predict([[7.5]]))
```
