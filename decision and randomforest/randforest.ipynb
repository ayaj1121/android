```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt
pos_salary=pd.read_csv("C:\\Users\\memon\\Documents\\python\\decision and randomforest\\Position_Salaries.csv")
X=pos_salary.iloc[:,1:2].values
y=pos_salary.iloc[:,2:3].values
from sklearn.ensemble import RandomForestRegressor
regresser=RandomForestRegressor(n_estimators=100,random_state=0)
regresser.fit(X,y)
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color="red")
plt.plot(X_grid,regresser.predict(X_grid),color="green")
plt.show()
```
