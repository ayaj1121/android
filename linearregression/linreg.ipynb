```python
import pandas as pd,numpy as np,matplotlib.pyplot as mp
saldat=pd.read_csv("C:\\Users\\memon\\Documents\\python\\linearregression\\Salary_Data.csv")
X=saldat.loc[:,["YearsExperience"]].values
y=saldat.loc[:,["Salary"]].values
from sklearn.model_selection  import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
regresser=LinearRegression()
regresser.fit(X_train,y_train)
y_pred=regresser.predict(X_test)
mp.scatter(X_train,y_train,color="red")
mp.scatter(X_test,y_test,color="blue")
mp.plot(X_train,regresser.predict(X_train),color="green")
mp.show()
print(regresser.coef_)
print(regresser.intercept_)
```
