```python
import pandas as pd,numpy as np,matplotlib.pyplot as mp
posit_salary=pd.read_csv("C:\\Users\\memon\\Documents\\python\\polynomialregression\\Position_Salaries.csv")
X=posit_salary.iloc[:,1:2].values
y=posit_salary.iloc[:,2:3].values
# mp.scatter(X,y,color="red")
from sklearn.preprocessing import PolynomialFeatures
polr_reg=PolynomialFeatures(degree=4)
X_Poly=polr_reg.fit_transform(X)
polr_reg.fit(X_Poly,y)
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression() 
lin_reg.fit(X_Poly,y)
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
mp.title("with X_grid")
mp.scatter(X,y,color="red")
mp.plot(X_grid,lin_reg.predict(polr_reg.fit_transform(X_grid)),color="green")
mp.show()
mp.title("with X")
mp.scatter(X,y,color="red")
mp.plot(X,lin_reg.predict(polr_reg.fit_transform(X)),color="green")
mp.show()
```
